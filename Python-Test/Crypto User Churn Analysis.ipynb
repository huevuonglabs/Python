```python
# Calculate median values of all columns for churned and retained users
df.groupby('label').median(numeric_only=True)
```

```python
#Snapshot Insights:
# 1. Engagement and Frequency:
# Retained users executed 9 transactions over 5 trading days, and opened apps on 10 days.
# Churen users executed only 4 transactions over 2 trading days, with app activity on just 4 days.
# Retained users have more individuals sessions (23), more than twice churned users (11)
# Retained users spend more time on app than churned users (203 mins vs. 101 mins)
# It means retained users engage more consistently, active on more days, open app more often, spend more time in the app.
# 2. Tenure
# Retained users have been onboarded for ~115 days, compared to only ~57 days for churned users.
# Longer tenure associated with higher retention
# 3. Trading Behavior & Volume
# Retained users trade nearly x2 total volume of churned users ($559 vs $296)
# Average trade size is nearly the same in two groups (~$62)
# Retained users trade a broader set of assets (5 assets vs 4 for churned users)
# It implies retained users are unnecessarily trading larger positions per trade but rather trading more frequently and consistently.
# 4. Advanced Features
# The median value for used_advanced_features is 0 for both groups, suggesting that while advanced features matter at the margin, most median users are spot traders. So differences in churn are not driven by advanced feature usage.
```

```python
#Deep Dive Analysis
```

```python
# Trade volume per trade day
df['volume_per_trade_day']= df['total_trade_volume_usd'] / df['trade_days']
median_volume_per_trade_day=df.groupby('label').median(numeric_only=True)[['volume_per_trade_day']]
median_volume_per_trade_day
```

```python
# Trades per activity day
df['trade_per_active_day']=df['transactions'] / df['activity_days']
median_trade_per_active_day=df.groupby('label').median(numeric_only=True)[['trade_per_active_day']]
median_trade_per_active_day
```

```python
# Minutes per session
df['minutes_per_session']=df['session_minutes_total'] / df['sessions']
median_minutes_per_session=df.groupby('label').median(numeric_only=True)[['minutes_per_session']]
median_minutes_per_session
```

```python
# Fees per $ volume
df['fee_per_volume']=df['fees_paid_usd'] / df['total_trade_volume_usd']
median_fee_per_volume=df.groupby('label').median(numeric_only=True)[['fee_per_volume']]
median_fee_per_volume
```

```python
# Medians of these derived metrics by churn segment
df.groupby('label').median(numeric_only=True)[['volume_per_trade_day','trade_per_active_day', 'minutes_per_session', 'fee_per_volume']] 
```

```python
# Churned users trades ~2.5x more volume per trading day than retain users ($318 vs$128). That implied churned users concentrate more capital into fewer trading days, suggesting short-term, high-intensity trading behaviorlikely speculation-driven rather than habit-driven. 
# Churned users execute slightly more trades per day when they open the app. When churned users are active, they are marginally more aggressive, compressing activity into short bursts instead of spreading it across many days
# Time per Session, fee per volume is nearly identical across both groups. Churn is not explained by time spent per visit or fee structure.
```
